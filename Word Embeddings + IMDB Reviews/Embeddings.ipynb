```python
import tensorflow_datasets as tfds
```

```python
imdb, info = tfds.load('imdb_reviews', with_info=True, as_supervised=True)
```

```python
import numpy as np

train_data,test_data = imdb['train'], imdb['test']
```

```python
train_sentences=[]
test_sentences=[]

train_labels=[]
test_labels=[]

for s,l in train_data:
    train_sentences.append(str(s.numpy()))
    train_labels.append(l.numpy())
    
for s,l in test_data:
    test_sentences.append(str(s.numpy()))
    test_labels.append(l.numpy())

#0: negative
#1: positive
```

```python
print(train_sentences[0])
print(train_labels[0])
```

```python
# nen mang thanh mang array

train_labels = np.array(train_labels)
test_labels = np.array(test_labels)
```

```python
vocab_size=10000
embedding_dim=64
max_length=140
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer(num_words=vocab_size,oov_token='<OOV>')
tokenizer.fit_on_texts(train_sentences)
```

```python
tokenizer.word_index
```

```python
# Gan nguoc lai train_sequence thanh cac so theo nhu da token ben tren

train_sequences = tokenizer.texts_to_sequences(train_sentences)
```

```python
# Cau nao dai thi cat o dang sau, cau nao ngan thi pad o dnag truoc

padded_train_sequences = pad_sequences(train_sequences,maxlen=max_length,truncating='post',padding='pre')
```

```python
padded_train_sequences
```

```python
padded_train_sequences.shape
```

```python
test_sequences = tokenizer.texts_to_sequences(test_sentences)
```

```python
test_sequences[0]
```

```python
padded_test_sequences = pad_sequences(test_sequences,maxlen=max_length,truncating='post',padding='post')
```

```python
padded_test_sequences.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Flatten, Dense
```

```python
model = Sequential()
model.add(Embedding(vocab_size, embedding_dim, input_length=max_length))
model.add(Flatten())
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['acc'])
```

```python
model.summary()
```

```python
model.fit(padded_train_sequences, train_labels, epochs=30, validation_data=(padded_test_sequences,test_labels))
```

```python
model.save('ReviewIMDB.h5')
```
